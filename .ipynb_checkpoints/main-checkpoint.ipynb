## 2. Now write a function that returns the largest element on a list

```python
def greatest(Numlist):
    if not Numlist:
        return None
    largest = Numlist[0]
    for i in Numlist:
        if i > largest:
            largest = i
    return largest

list_example = [4,1,80,20,100,1,150]
greatest(list_example)
```

```python
# This will test your function 
test_greatest(greatest)
```

## 5. Now combine those two ideas and write a function that receives a list and either "+" or "*" and outputs acordingly

```python
def oper_all(arr, oper = "*"):
    if oper == "*":
        if not arr:
            return 1
        return arr[0] * oper_all(arr[1:], "*")
    elif oper == "+":
        if not arr:
            return 0
        return arr[0] + oper_all(arr[1:], "+")

nums = [2,3,5]
print(oper_all(nums, oper = "*"))
print(oper_all(nums, oper = "+"))
```

```python
# This will test your function 
test_operations(oper_all)
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    if n<=0:
        return 1
    else:
        return n * factorial(n-1)

factorial(5)
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

```python
# This will test your function 
test_factorial(factorial)
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. 🤔`

```python
def unique(lst_un):
    unique_list = []
    for i in lst_un:
        if i not in unique_list:
            unique_list.append(i)
    return unique_list

list_example = [3,3,4,6,7,5,4,10,9,10]
unique(list_example)
```

```python
# This will test your function 
test_unique(unique)
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... 🧐`

```python
def mode_counter(arr):
    if not arr:
        return None
    
    counts = {}
    for item in arr:
        try:
            counts[item] += 1
        except KeyError:
            counts[item] = 1
    
    mode_value = None
    max_count = 0

    for key, count in counts.items():
        if count > max_count:
            max_count = count
            mode_value = key

    return mode_value

data = [1, 3, 3, 3, 2, 1, 1, 3, 2, 4]
print(f"The mode is: {mode_counter(data)}")
```

```python
# This will test your function 
test_mode(mode_counter)
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. 😉`

```python
def st_dev(list_sd):
    count = 0
    total = 0
    for num in list_sd:
        count += 1
        total += num
        
    if count <= 1:
        return 0.0
        
    mean = total / count
    
    sq_diff_sum = 0
    for num in list_sd:
        sq_diff_sum += (num - mean) ** 2
        
    # Calculate Variance 
    # CRITICAL: Use (count - 1) for Sample Standard Deviation as the built-in function in python does
    variance = sq_diff_sum / (count - 1)
    
    # 5. Return Square Root using the exponent operator
    return variance ** 0.5

data = [1,2,3,4,5]
st_dev(data)
```

```python
# This will test your function 
test_stdev(st_dev)
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
    found_letters = {}
    for char in string:
        lower_char = char.lower()
        if 'a' <= lower_char <= 'z':
            found_letters[lower_char] = True

    alphabet = "abcdefghijklmnopqrstuvwxyz"
    
    for letter in alphabet:
        try:
            found_letters[letter]
        except KeyError:
            return False
    
    return True
            
# Test cases
print(pangram("The quick brown fox jumps over the lazy dog")) # True
print(pangram("Hello World"))                                 # False
```

```python
# This will test your function 
test_pangram(pangram)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! 🤪`

```python
def sort_alpha(string):
    # Manual "Split": Convert string to a list
    words_list = []
    current_word = ""
    
    for char in string + "," : # The + "," only happens once at the very end of the entire string. It doesn't put a comma after every word
        if char == ",":
            words_list.append(current_word)
            current_word = ""
        else:
            current_word += char
    #print(words_list)
    
    # Sorting
    sorted_words = sorted(words_list)
    #print(sorted_words)
    
    # Manual "Join": Convert list back to a string
    result_string = ""
    first_word = True
    
    for word in sorted_words:
        if first_word:
            result_string = word
            first_word = False
        else:
            # For every word after the first, add a comma before it
            result_string = result_string + "," + word
            
    return result_string

print(sort_alpha("banana,apple,cherry,date")) 
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
    length = 0
    has_lower = False
    has_upper = False
    has_digit = False
    has_special = False

    for char in password:
        length += 1  # Manually counting the length
        
        if 'a' <= char <= 'z':
            has_lower = True
        elif 'A' <= char <= 'Z':
            has_upper = True
        elif '0' <= char <= '9':
            has_digit = True
        else:
            # If it's not a letter or a number, it's a special character
            has_special = True

    if length >= 8 and has_lower and has_upper and has_digit and has_special:
        return True
    else:
        return False

print(check_pass("Abc123!@")) 
print(check_pass("password"))
```

```python
# This will test your function 
test_pass(check_pass)
```
